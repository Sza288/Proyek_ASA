```python
import heapq
import time
import math
import pandas as pd
import numpy as np

# ==========================================
# FUNGSI HEURISTIK & UTALITAS
# ==========================================

def heuristik(G, node1, node2):
  """
  Menghitung jarak lingkaran besar antara dua node menggunakan Formula Haversine asli berdasarkan koordinat Lintang (y) dan Bujur (x).
  """

  lon1, lat1 = G.nodes[node1]['x'], G.nodes[node1]['y']
  lon2, lat2 = G.nodes[node2]['x'], G.nodes[node2]['y']
  R = 6371000.0
  phi1 = math.radians(lat1)
  phi2 = math.radians(lat2)
  delta_phi = math.radians(lat2 - lat1)
  delta_lambda = math.radians(lon2 - lon1)

  a = (math.sin(delta_phi / 2.0) ** 2 + math.cos(phi1) * math.cos(phi2) * (math.sin(delta_lambda / 2.0) ** 2))

  c = 2.0 * math.atan2(math.sqrt(a), math.sqrt(1.0 - a))
  jarak_meter = R * c
  return jarak_meter

def check_line_of_sight(G, node1, node2):
  """
  LoS dianggap valid jika ada koneksi langsung ATAU jarak udara antar titik sangat dekat (<= 80 meter), merepresentasikan turis yang bisa memotong area trotoar.
  """
  if G.has_edge(node1, node2) or G.has_edge(node2, node1):
    return True

  # Jika tidak ada edge langsung, cek jarak geometri lurusnya
  jarak_pandang = heuristik(G, node1, node2)
  if jarak_pandang <= 80.0: # Batas radius pandang mata pejalan kaki
    return True

  return False

def format_rute(rute_idx, daftar_nama):
  """Mengubah urutan indeks integer menjadi urutan nama lokasi terformat."""
  return " ➔ ".join([daftar_nama[idx] for idx in rute_idx])

# ==========================================
# FASE 1: IMPLEMENTASI ROUTE PLANNING
# ==========================================

def greedy_best_first_search(G, start, goal):
  pq = []
  h_start = heuristik(G, start, goal)
  heapq.heappush(pq, (h_start, start))

  visited = set()
  parent = {start: None}
  nodes_expanded = 0

  while pq:
    h_score, current = heapq.heappop(pq)
    nodes_expanded += 1

    if current == goal:
      total_jarak_riil = 0
      curr = goal
      while parent[curr] is not None:
        p = parent[curr]
        edge_data = G.get_edge_data(p, curr)
        weight = edge_data[0].get('length', 1) if edge_data else 1
        total_jarak_riil += weight
        curr = p
      return total_jarak_riil, nodes_expanded

    if current in visited:
      continue
    visited.add(current)

    for neighbor in G.neighbors(current):
      if neighbor not in visited and neighbor not in parent:
        parent[neighbor] = current
        h_score_neighbor = heuristik(G, neighbor, goal)
        heapq.heappush(pq, (h_score_neighbor, neighbor))

  return float('inf'), nodes_expanded

def uniform_cost_search(G, start, goal):
  pq = []
  heapq.heappush(pq, (0, start))
  cost_so_far = {start: 0}
  nodes_expanded = 0

  while pq:
    current_cost, current = heapq.heappop(pq)
    nodes_expanded += 1

    if current == goal:
      return current_cost, nodes_expanded

    if current_cost > cost_so_far.get(current, float('inf')):
      continue

    for neighbor in G.neighbors(current):
      edge_data = G.get_edge_data(current, neighbor)
      weight = edge_data[0].get('length', 1) if edge_data else 1
      new_cost = current_cost + weight

      if new_cost < cost_so_far.get(neighbor, float('inf')):
        cost_so_far[neighbor] = new_cost
        heapq.heappush(pq, (new_cost, neighbor))

  return float('inf'), nodes_expanded

def a_star_search(G, start, goal):
  pq = []
  h_start = heuristik(G, start, goal)
  heapq.heappush(pq, (h_start, 0, start)) # (f_score, g_score, node)
  cost_so_far = {start: 0}
  nodes_expanded = 0

  while pq:
    _, current_cost, current = heapq.heappop(pq)
    nodes_expanded += 1

    if current == goal:
      return current_cost, nodes_expanded

    if current_cost > cost_so_far.get(current, float('inf')):
      continue

    for neighbor in G.neighbors(current):
      edge_data = G.get_edge_data(current, neighbor)
      weight = edge_data[0].get('length', 1) if edge_data else 1
      new_cost = current_cost + weight

      if new_cost < cost_so_far.get(neighbor, float('inf')):
        cost_so_far[neighbor] = new_cost
        f_score = new_cost + heuristik(G, neighbor, goal)
        heapq.heappush(pq, (f_score, new_cost, neighbor))

  return float('inf'), nodes_expanded

def theta_star_search(G, start, goal):
  pq = []
  h_start = heuristik(G, start, goal)
  heapq.heappush(pq, (h_start, 0, start))
  cost_so_far = {start: 0}
  parent = {start: start}
  nodes_expanded = 0

  while pq:
    _, current_cost, current = heapq.heappop(pq)
    nodes_expanded += 1

    if current == goal:
      return current_cost, nodes_expanded

    if current_cost > cost_so_far.get(current, float('inf')):
      continue

    for neighbor in G.neighbors(current):
      curr_parent = parent[current]

      # Aturan Utama Theta*: Cek Jalur Garis Pandang (Path 2)
      if check_line_of_sight(G, curr_parent, neighbor):
        los_weight = heuristik(G, curr_parent, neighbor)
        new_cost = cost_so_far[curr_parent] + los_weight
        if new_cost < cost_so_far.get(neighbor, float('inf')):
          cost_so_far[neighbor] = new_cost
          parent[neighbor] = curr_parent
          f_score = new_cost + heuristik(G, neighbor, goal)
          heapq.heappush(pq, (f_score, new_cost, neighbor))
      else:
        # A* Biasa (Path 1)
        edge_data = G.get_edge_data(current, neighbor)
        weight = edge_data[0].get('length', 1) if edge_data else 1
        new_cost = current_cost + weight
        if new_cost < cost_so_far.get(neighbor, float('inf')):
          cost_so_far[neighbor] = new_cost
          parent[neighbor] = current
          f_score = new_cost + heuristik(G, neighbor, goal)
          heapq.heappush(pq, (f_score, new_cost, neighbor))

  return float('inf'), nodes_expanded
```

```python
# =============================================================================
# MEMBANGUN MATRIX (15 x 15 LOKASI)
# =============================================================================
import numpy as np
import time
import pandas as pd
from IPython.display import display, Markdown

target_nodes = nodes_osm

# Inisialisasi matriks kosong untuk setiap algoritma
matriks_ucs = np.zeros((n_lokasi, n_lokasi))
matriks_gbfs = np.zeros((n_lokasi, n_lokasi))
matriks_astar = np.zeros((n_lokasi, n_lokasi))
matriks_theta = np.zeros((n_lokasi, n_lokasi))

# Penampung data performa pembuatan matriks (Fase 1)
waktu_ucs, ekspansi_ucs = 0, 0
waktu_gbfs, ekspansi_gbfs = 0, 0
waktu_astar, ekspansi_astar = 0, 0
waktu_theta, ekspansi_theta = 0, 0

# 1. Eksekusi UCS
t0 = time.time()
for i in range(n_lokasi):
  for j in range(n_lokasi):
    if i == j: continue
    dist, exp = uniform_cost_search(G, target_nodes[i], target_nodes[j])
    matriks_ucs[i][j] = dist
    ekspansi_ucs += exp
waktu_ucs = (time.time() - t0) * 1000

# 2. Eksekusi GBFS
t0 = time.time()
for i in range(n_lokasi):
  for j in range(n_lokasi):
    if i == j: continue
    dist, exp = greedy_best_first_search(G, target_nodes[i], target_nodes[j])
    matriks_gbfs[i][j] = dist
    ekspansi_gbfs += exp
waktu_gbfs = (time.time() - t0) * 1000

# 3. Eksekusi A*
t0 = time.time()
for i in range(n_lokasi):
  for j in range(n_lokasi):
    if i == j: continue
    dist, exp = a_star_search(G, target_nodes[i], target_nodes[j])
    matriks_astar[i][j] = dist
    ekspansi_astar += exp
waktu_astar = (time.time() - t0) * 1000

# 4. Eksekusi Theta*
t0 = time.time()
for i in range(n_lokasi):
  for j in range(n_lokasi):
    if i == j: continue
    dist, exp = theta_star_search(G, target_nodes[i], target_nodes[j])
    matriks_theta[i][j] = dist
    ekspansi_theta += exp
waktu_theta = (time.time() - t0) * 1000

# =============================================================================
# MENAMPILKAN MATRIKS
# =============================================================================
def tampilkan_matriks(mat, algoritma):
  df = pd.DataFrame(mat, columns=[f'L{i}' for i in range(n_lokasi)], index=[f'L{i}' for i in range(n_lokasi)])
  display(Markdown(f'### Matriks Jarak: {algoritma}'))
  display(df.round(1))

# Menampilkan matriks
tampilkan_matriks(matriks_ucs, 'Uniform Cost Search (UCS)')
tampilkan_matriks(matriks_gbfs, 'Greedy Best-First Search (GBFS)')
tampilkan_matriks(matriks_astar, 'A* Search')
tampilkan_matriks(matriks_theta, 'Theta* Search')
```

```python
# ─────────────────────────────────────────────────────────────────────────────
# FASE 2 — Dynamic Programming Held-Karp: TSP
# PERAN DALAM SISTEM 2-TAHAP:
#   Menerima Matriks Jarak N×N (output Fase 1) sebagai masukan.
#   Menggunakan representasi bitmask untuk merepresentasikan himpunan yang dikunjungi.
# ─────────────────────────────────────────────────────────────────────────────

def held_karp_dp(matriks, start_node=0):
  """
  INPUT  : matriks jarak N×N
  OUTPUT : (rute_optimal, total_jarak_minimal, jumlah_state_dievaluasi)
  """
  n = len(matriks)
  JUMLAH_MASK = 1 << n
  INF = float('inf')

  # dp[mask][i] = biaya minimum mencapai simpul i dengan subset dikunjungi = mask
  # Inisialisasi semua INF
  dp   = [[INF] * n for _ in range(JUMLAH_MASK)]
  prev = [[-1]  * n for _ in range(JUMLAH_MASK)]

  # Basis: mulai dari start_node
  dp[1 << start_node][start_node] = 0.0
  state_dievaluasi = 0

  # Bottom-up: iterasi semua subset (mask) dari kecil ke besar
  for mask in range(JUMLAH_MASK):
    for u in range(n):
      # Simpul u harus ada di dalam mask
      if not (mask & (1 << u)):
        continue
      if dp[mask][u] == INF:
        continue

      # Perluas ke simpul v yang belum dikunjungi
      for v in range(n):
        if mask & (1 << v):   # v sudah dikunjungi
          continue
        state_dievaluasi += 1
        new_mask = mask | (1 << v)
        biaya_baru = dp[mask][u] + matriks[u][v]
        if biaya_baru < dp[new_mask][v]:
          dp[new_mask][v] = biaya_baru
          prev[new_mask][v] = u

  # Semua node telah dikunjungi → kembali ke start_node
  mask_penuh = JUMLAH_MASK - 1
  jarak_minimal = INF
  last_node = -1
  for u in range(n):
    if u == start_node:
      continue
    total = dp[mask_penuh][u] + matriks[u][start_node]
    if total < jarak_minimal:
      jarak_minimal = total
      last_node = u

  # Rekonstruksi rute dari tabel prev
  rute = []
  mask = mask_penuh
  curr = last_node
  while curr != -1:
    rute.append(curr)
    prev_node = prev[mask][curr]
    mask ^= (1 << curr)
    curr = prev_node
  rute.reverse()
  rute.append(start_node)   # kembali ke titik awal

  return rute, jarak_minimal, state_dievaluasi
```

```python
# ─────────────────────────────────────────────────────────────────────────────
# EKSEKUSI SISTEM (VARIASI N = 5, 10, 15)
# ─────────────────────────────────────────────────────────────────────────────

pd.set_option('display.max_colwidth', None)

# Daftar variasi N
variasi_n = [5, 10, 15]
# List untuk menampung seluruh hasil dari semua variasi N
semua_hasil = []

for n_uji in variasi_n:
  daftar_sistem = [
    {"nama": "Greedy BFS + DP", "matriks": matriks_gbfs[:n_uji, :n_uji],  "waktu_fase1": waktu_gbfs,  "ekspansi_fase1": ekspansi_gbfs},
    {"nama": "UCS + DP",        "matriks": matriks_ucs[:n_uji, :n_uji],   "waktu_fase1": waktu_ucs,   "ekspansi_fase1": ekspansi_ucs},
    {"nama": "A* + DP",         "matriks": matriks_astar[:n_uji, :n_uji], "waktu_fase1": waktu_astar, "ekspansi_fase1": ekspansi_astar},
    {"nama": "Theta* + DP",     "matriks": matriks_theta[:n_uji, :n_uji], "waktu_fase1": waktu_theta, "ekspansi_fase1": ekspansi_theta}
  ]

  hasil_per_n = []
  names_uji = names[:n_uji]

  for sistem in daftar_sistem:
    mat_f1 = sistem["matriks"]
    t0_dp = time.time()
    rute, jarak_total, state_dp = held_karp_dp(mat_f1, start_node=0)
    waktu_fase2 = (time.time() - t0_dp) * 1000

    waktu_total = sistem["waktu_fase1"] + waktu_fase2

    row = {
      "Sistem": sistem["nama"],
      "Jumlah Lokasi (N)": n_uji,
      "Rute Optimal": format_rute(rute, names_uji),
      "Jarak (m)": round(jarak_total, 1),
      "Waktu Fase 1 (ms)": round(sistem["waktu_fase1"], 1),
      "Waktu Fase 2 (ms)": round(waktu_fase2, 1),
      "Total Waktu Eksekusi (ms)": round(waktu_total, 1),
      "Ekspansi Simpul": sistem["ekspansi_fase1"],
      "Ekspansi Status DP": state_dp
    }
    hasil_per_n.append(row)
    semua_hasil.append(row)

  df_tabel = pd.DataFrame(hasil_per_n)
  display(Markdown(f'### Tabel Perbandingan Sistem (N={n_uji})'))
  display(df_tabel)
  print("\n")

df_sistem = pd.DataFrame(semua_hasil)
```
